```python
""" 1. Given a 1-indexed array of integers numbers that are already sorted in non-decreasing order, find two numbers such that they add up to a specific target number. Let these two numbers be numbers[index1] and numbers[index2] where 1 <= index1 < index2 < numbers.length.
Return the indices of the two numbers, index1, and index2, added by one as an integer array [index1, index2] of length 2.
The tests are generated such that there is exactly one solution. You may not use the same element twice.
Your solution must use only constant extra space.
Input: numbers = [2,7,11,15], target = 9
Output: [1,2]
Explanation: The sum of 2 and 7 is 9. 
Therefore, index1 = 1, index2 = 2. 
We return [1, 2]."""

#Answer

def two_sum(numbers, target):
    left, right = 0, len(numbers) - 1

    while left < right:
        current_sum = numbers[left] + numbers[right]

        if current_sum == target:
            return [left + 1, right + 1]  # Adding 1 to the indices to make them 1-indexed
        elif current_sum < target:
            left += 1
        else:
            right -= 1

    # If no solution is found, you may return an empty array or raise an error depending on the requirements.
    return []

# Test the function with the given example

numbers = [2, 7, 11, 15]
target = 9
result = two_sum(numbers, target)
print(result)  
```

```python
""" 2. Given an array of integer nums sorted in non-decreasing order, find the starting and ending position of a given target value.
If the target is not found in the array, return [-1, -1].
You must write an algorithm with O(log n) runtime complexity
Input: nums = [5,7,7,8,8,10], target = 8
Output: [3,4]
"""

#Answer

def find_first_occurrence(nums, target):
    left, right = 0, len(nums) - 1
    result = -1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] >= target:
            right = mid - 1
            if nums[mid] == target:
                result = mid
        else:
            left = mid + 1

    return result

def find_last_occurrence(nums, target):
    left, right = 0, len(nums) - 1
    result = -1

    while left <= right:
        mid = left + (right - left) // 2

        if nums[mid] <= target:
            left = mid + 1
            if nums[mid] == target:
                result = mid
        else:
            right = mid - 1

    return result

def search_range(nums, target):
    first_occurrence = find_first_occurrence(nums, target)
    last_occurrence = find_last_occurrence(nums, target)

    return [first_occurrence, last_occurrence]

# Test the function with the given example
nums = [5, 7, 7, 8, 8, 10]
target = 8
result = search_range(nums, target)
print(result)
```

```python
''' 3. A peak element is an element that is strictly greater than its neighbors.
Given a 0-indexed integer array nums, find a peak element, and return its index. If the array contains multiple peaks, return the index to any of the peaks.
You may imagine that nums[-1] = nums[n] = -âˆž. In other words, an element is always considered to be strictly greater than a neighbor that is outside the array.
You must write an algorithm that runs in O(log n) time.
Input: nums = [1,2,3,1]
Output: 2
Explanation: 3 is a peak element and your function should return the index number 2.'''

# Answer

def find_peak_element(nums):
    left, right = 0, len(nums) - 1

    while left < right:
        mid = left + (right - left) // 2

        if nums[mid] < nums[mid + 1]:
            left = mid + 1
        else:
            right = mid

    return left

# Test the function with the given example
nums = [1, 2, 3, 1]
result = find_peak_element(nums)
print(result) 
```

```python
''' 4. Given a sorted array of distinct integers and a target value, return the index if the target is found. If not, return the index where it would be if it were inserted in order.
You must write an algorithm with O(log n) runtime complexity.
Input: nums = [1,3,5,6], target = 5
Output: 2

Input: nums = [1,3,5,6], target = 7
Output: 4
'''

# Answer

def search_insert(nums, target):
    left, right = 0, len(nums)

    while left < right:
        mid = left + (right - left) // 2

        if nums[mid] < target:
            left = mid + 1
        else:
            right = mid

    return left

# Test the function with the given examples
nums1 = [1, 3, 5, 6]
target1 = 5
result1 = search_insert(nums1, target1)
print(result1)  

nums2 = [1, 3, 5, 6]
target2 = 7
result2 = search_insert(nums2, target2)
print(result2)  
```

```python
''' 5. Find the majority element in the array. A majority element in an array A[] of size n is an element that appears more than n/2 times (and hence there is at most one such element). 
Input: A[]={3, 3, 4, 2, 4, 4, 2, 4, 4}
Output: 4
Explanation: The frequency of 4 is 5 which is greater than half of the size of the array size. '''

#Answer

def find_majority_element(nums):
    candidate = None
    count = 0

    for num in nums:
        if count == 0:
            candidate = num
            count = 1
        elif num == candidate:
            count += 1
        else:
            count -= 1

    # At this point, the candidate is the potential majority element
    # We need to verify if it indeed appears more than n/2 times in the array
    count = 0
    for num in nums:
        if num == candidate:
            count += 1

    if count > len(nums) // 2:
        return candidate
    else:
        return None  # There is no majority element

# Test the function with the given example
nums = [3, 3, 4, 2, 4, 4, 2, 4, 4]
result = find_majority_element(nums)
print(result)  
```

```python
''' 7. Given an array of integers, find the inversion of an array. Formally, two elements a[i] and a[j] form an inversion if a[i] > a[j] and i < j.
N=5, arr[] = {2, 4, 1, 3, 5}
Output: 3
Explanation: (2,1) (4,1) and (4,3) forms an inversion in an array.'''

# Answer

def merge_and_count(arr, left, mid, right):
    count = 0
    temp = [0] * (right - left + 1)
    i, j, k = left, mid + 1, 0

    while i <= mid and j <= right:
        if arr[i] <= arr[j]:
            temp[k] = arr[i]
            i += 1
        else:
            temp[k] = arr[j]
            count += (mid - i + 1)  # Count the inversions
            j += 1
        k += 1

    while i <= mid:
        temp[k] = arr[i]
        i += 1
        k += 1

    while j <= right:
        temp[k] = arr[j]
        j += 1
        k += 1

    for i in range(len(temp)):
        arr[left + i] = temp[i]

    return count

def merge_sort_and_count(arr, left, right):
    count = 0

    if left < right:
        mid = left + (right - left) // 2
        count += merge_sort_and_count(arr, left, mid)
        count += merge_sort_and_count(arr, mid + 1, right)
        count += merge_and_count(arr, left, mid, right)

    return count

def inversion_count(arr):
    return merge_sort_and_count(arr, 0, len(arr) - 1)

# Test the function with the given example
arr = [2, 4, 1, 3, 5]
result = inversion_count(arr)
print(result)
```

```python
''' 8. Given three arrays sorted in non-decreasing order, print all common elements in these arrays.
ar1[] = {1, 5, 10, 20, 40, 80} 
ar2[] = {6, 7, 20, 80, 100} 
ar3[] = {3, 4, 15, 20, 30, 70, 80, 120} 
Output: 20, 80
'''

# Answer

def find_common_elements(ar1, ar2, ar3):
    common_elements = []
    i, j, k = 0, 0, 0

    while i < len(ar1) and j < len(ar2) and k < len(ar3):
        if ar1[i] == ar2[j] == ar3[k]:
            common_elements.append(ar1[i])
            i += 1
            j += 1
            k += 1
        elif ar1[i] <= ar2[j] and ar1[i] <= ar3[k]:
            i += 1
        elif ar2[j] <= ar1[i] and ar2[j] <= ar3[k]:
            j += 1
        else:
            k += 1

    return common_elements

# Test the function with the given example
ar1 = [1, 5, 10, 20, 40, 80]
ar2 = [6, 7, 20, 80, 100]
ar3 = [3, 4, 15, 20, 30, 70, 80, 120]
result = find_common_elements(ar1, ar2, ar3)
print(result) 
```
